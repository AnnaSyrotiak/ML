1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = 10*x - 5
df = sp.diff(f, x)

print(f"Функція: {f}")
print(f"Похідна: {df}")
```

```python
f_func = sp.lambdify(x, f, modules='numpy')
df_func = sp.lambdify(x, df, modules='numpy')

x_vals = np.linspace(0, 10, 100)
y_vals = f_func(x_vals)

dy_vals = [float(df) for _ in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5 (Оцінка)', color='blue')

plt.plot(x_vals, dy_vals, label="f'(x) = 10 (Похідна)", color='green', linestyle='--')

plt.title("Залежність оцінки від годин навчання")
plt.xlabel("Години навчання (x)")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)
plt.show()
```

Кожна додаткова година навчання збільшує оцінку рівно на 10 балів. Цей приріст є постійним і не залежить від того, скільки вже тривало навчання

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
df1 = sp.diff(f1, x)

print(f"1. Функція: {f1}")
print(f"   Похідна: {df1}")

f2 = sp.ln(x)
df2 = sp.diff(f2, x)

print(f"2. Функція: {f2}")
print(f"   Похідна: {df2}")

f1_func = sp.lambdify(x, f1, modules='numpy')
df1_func = sp.lambdify(x, df1, modules='numpy')
f2_func = sp.lambdify(x, f2, modules='numpy')
df2_func = sp.lambdify(x, df2, modules='numpy')

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

x_vals_sin = np.linspace(-2*np.pi, 2*np.pi, 400)
ax[0].plot(x_vals_sin, f1_func(x_vals_sin), label='f(x) = sin(x)', color='blue')
ax[0].plot(x_vals_sin, df1_func(x_vals_sin), label="f'(x) = cos(x)", color='orange', linestyle='--')
ax[0].set_title("Функція sin(x) та її похідна")
ax[0].legend()
ax[0].grid(True)

x_vals_ln = np.linspace(0.1, 10, 400)
ax[1].plot(x_vals_ln, f2_func(x_vals_ln), label='f(x) = ln(x)', color='green')
ax[1].plot(x_vals_ln, df2_func(x_vals_ln), label="f'(x) = 1/x", color='red', linestyle='--')
ax[1].set_title("Функція ln(x) та її похідна")
ax[1].legend()
ax[1].grid(True)

plt.show()
```

1. Для $sin(x)$:

  Похідна дорівнює $cos(x)$.На графіку видно, що коли синус зростає (лінія йде вгору), косинус додатний (вище нуля). Коли синус досягає піку (вершини), косинус дорівнює нулю.

2. Для $ln(x)$:

  Похідна дорівнює $1/x$.На графіку видно, що логарифм постійно зростає, але з часом все повільніше. Це підтверджує графік похідної (червоний пунктир): він завжди додатний (бо функція росте), але стрімко спадає до нуля (бо швидкість росту падає).


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)

print(f"Функція витрат: {f}")
print(f"Похідна : {df}")

f_func = sp.lambdify(x, f, modules='numpy')
df_func = sp.lambdify(x, df, modules='numpy')

x_vals = np.linspace(0, 100, 200)
y_vals = f_func(x_vals)
dy_vals = df_func(x_vals)


plt.figure(figsize=(12, 6))

plt.plot(x_vals, y_vals, label='f(x) = $x^2 - 5x + 1$ (Загальні витрати)', color='blue')

# Графік похідної
plt.plot(x_vals, dy_vals, label="f'(x) = $2x - 5$ (Похідна)", color='orange', linestyle='--')

plt.title("Загальні витрати та похідна")
plt.xlabel("Обсяг виробництва (x)")
plt.ylabel("Вартість")
plt.legend()
plt.grid(True)
plt.show()
```
