**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np

list_array = np.array([5, 15, 25, 35, 45])
print("Масив зі списку:", list_array)

zero_array = np.zeros(7)
print("Масив з нулів:", zero_array)

ones_array = np.ones(6)
print("Масив з одиниць:", ones_array)

range_array = np.arange(10,19)
print("Масив з числами від 10 до 19:", range_array)
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
import numpy as np

list_array = np.array([12, 24, 36, 48, 60])

double = list_array * 2
print("Масив, помножений на 2:", double)

plus_ten = list_array + 10
print("Масив, до якого додано 10:", plus_ten)

squared = list_array ** 2
print("Масив, піднесений до квадрату:", squared)

sqrt_array = np.sqrt(list_array)
print("Масив, квадратний корінь:", sqrt_array)
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
import numpy as np

data = np.array ([18, 27, 33, 41, 52, 68, 79])
data_avg = data.mean()
print("Середнє значення:", data_avg)

data_min = data.min()
print("Мінімум:", data_min)

data_max = data.max()
print("Максимум:", data_max)

data_sum = data.sum()
print("Сума всіх елементів:", data_sum)
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
import numpy as np

arr = np.array([15, 22, 35, 44, 58, 67])
first_element = arr[0]
print("Перший елемент:", first_element)

last_element = arr[arr.size-1]
print("Останній елемент:", last_element)

sliced_elements = arr[1:4]
print("Елементи з індексами 1 до 3:", sliced_elements)

elements_greater_than_40 = arr[arr > 40]
print("Елементи більші за 40:", elements_greater_than_40)
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів
- Обчисліть середнє всіх елементів

```python
import numpy as np

matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])

print("Форма матриці (shape):", np.shape(matrix))
print("Перший рядок:", matrix[0])
print("Друга колонка:", matrix[:, 1])
print("Сума всіх елементів:", np.sum(matrix))
print("Середнє всіх елементів:", np.mean(matrix))
```
