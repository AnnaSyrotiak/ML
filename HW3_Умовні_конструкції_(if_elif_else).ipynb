
**1. Основи роботи з умовними конструкціями (if-elif-else)**


```python
cooking_time = 5
if cooking_time < 4:
    print("Яйце рідке")
elif 4 <= cooking_time < 7:
    print("Яйце всмятку")
elif 7 <= cooking_time < 10:
    print("Яйце вкруту")
else:
    print("Яйце переварене")
```

**2. Перевірка оцінки**

```python
grade = 72
if grade>=90:
    print("Відмінно")
elif 75 <= grade < 90:
    print ("Добре")
elif 60 <= grade < 75:
    print ("Задовільно")
else:
    print ("Незадовільно")
```

**3. Перевірка на наявність**

```python
food = ["молоко", "яйця", "масло", "сир"]
if "яйця" in food:
  print("Яйця є в холодильнику")
else:
  print("Яєць немає")

result = "Хліб є в холодильнику" if "хліб" in food else "Хліба немає"
print(result)
```
