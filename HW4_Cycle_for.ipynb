**1. Цикл for - базовий**

```python
for i in range (1, 8):
    print(i)
```

**2. Цикл for по списку**

```python
names = ["Олег", "Анна", "Іван", "Марія"]
for name in names:
    print(f"Привіт, {name}!")
```

**3. Цикл з enumerate**

```python
grades = [85, 92, 78, 95, 88]
for i,grade in enumerate (grades):
  print(f"Студент {i+1}: {grade} балів ")
```

**4. Сума парних чисел**

Напишіть програму, яка обчислює суму всіх парних чисел від 1 до 20.

```python
total_sum = 0
for i in range(1, 21):
    if i % 2 == 0:
        total_sum += i
print(f"Сума парних чисел від 1 до 20: {total_sum}")
```
