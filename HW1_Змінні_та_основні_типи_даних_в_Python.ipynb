**2. Арифметичні операції**


```python
var1 = 15
var2 = 4
print(f'Сума: {var1 + var2}')
print(f'Різниця: {var1 - var2}')
print(f'Добуток: {var1 * var2}')
print(f'Ділення: {var1 / var2}')
print(f'Цілочисельне ділення: {var1 // var2}')
print(f'Залишок від ділення: {var1 % var2}')
print(f'Піднесення до степеня: {var1 ** var2}')
```
